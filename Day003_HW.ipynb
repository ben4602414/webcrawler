### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


```python
# 讀檔案_File I/O
fh = open("./example/sample.xml", "r")
xml = fh.read()
fh.close()

print(xml)

# 讀檔案_xmltodict
import xmltodict

with open('./example/sample.xml') as fd:
    doc = dict(xmltodict.parse(fd.read()))

# 存取我們的資訊
print(doc)
print(doc['CUPOY']['Title'])

# 用迴圈存取我們的資訊
chapters = doc['CUPOY']['Chapters']['Chapter']
for chapter in chapters:
    print (chapter['@name'], chapter['#text'])
```

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

with open('./example/64_72hr_CH.xml') as fd:
    doc = dict(xmltodict.parse(fd.read()))

locations = doc['cwbopendata']['dataset']['locations']['location']
counts = 0

for location in locations:
    print(location['locationName'])
    counts += 1
print(counts)
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

with open('./example/64_72hr_CH.xml') as fd:
    doc = dict(xmltodict.parse(fd.read()))

locations = doc['cwbopendata']['dataset']['locations']['location']

for location in locations:
    
    #相同的標籤有很多個的時候，會以list的型態儲存
    
    datetime = location['weatherElement'][0]['time'][0]['dataTime']
    temperature_value = location['weatherElement'][0]['time'][0]['elementValue']['value']
    temperature_unit = location['weatherElement'][0]['time'][0]['elementValue']['measures']
    temperature = temperature_value+temperature_unit
    
    print(location['locationName'])
    print(datetime)
    print(temperature)
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

with open('./example/64_72hr_CH.xml') as fd:
    doc = dict(xmltodict.parse(fd.read()))

locations = doc['cwbopendata']['dataset']['locations']['location'][0]
location_1st = locations['locationName']


for x in locations['weatherElement'][0]['time']:
        
    datetime = x['dataTime']
    temperature_value = x['elementValue']['value']
    temperature_unit = x['elementValue']['measures']
    temperature = temperature_value+temperature_unit
    
    print(location_1st)
    print(datetime)
    print(temperature)
```
