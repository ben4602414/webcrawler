### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find('a').text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
#找到儲存類別分類的html
category = []
for d in soup.find(class_="part_list_2").find_all('h3'):
    if d.find('em').text not in category:
        category.append(d.find('em').text)
print(category)
```

### ③ 哪一個類別的文章最多

```python
#計算每一個類別的文章數量
counts = {}
for d in soup.find(class_="part_list_2").find_all('h3'):
    if d.find('em').text not in counts:
        counts[d.find('em').text] = 1
    else:
        counts[d.find('em').text] += 1
print(counts)

#找出文章最多的類別
value = 1
for c in counts:
    if counts[c] > value:
        most_chapters = c
        value = counts[c]
    elif counts[c] == value:
        most_chapters += "與" + str(c)

print("文章最多的類別為：" + most_chapters)
```
