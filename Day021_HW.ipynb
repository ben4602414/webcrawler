## 作業目標

根據範例：

1. 取出今天所有的新聞
2. 取出現在時間兩小時內的新聞
3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='/Users/ben/webcrawler/chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")
```

## 1. 取出今天所有的新聞

```python
import time
for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
import datetime as dt
now = dt.datetime.now()
delta = dt.timedelta(hours=2)
end_time = now - delta
print(now)
print(end_time)

for d in soup.find(class_="part_list_2").find_all('h3'):
    T = dt.datetime.strptime(d.find(class_='date').text, '%Y/%m/%d %H:%M')

    if T < end_time:
        break

    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
#定義前三天的日期及時間
now = dt.datetime.now()
minus_day = dt.timedelta(days=3)
last_3days = now - minus_day
pm_3 = dt.datetime(last_3days.year,last_3days.month,last_3days.day, 15, 0, 0)
pm_5 = dt.datetime(last_3days.year,last_3days.month,last_3days.day, 17, 0, 0)

#根據日期創造新的link
new_link = "https://www.ettoday.net/news/news-list-{}-{}-{}-0.htm".format(last_3days.year,last_3days.month,last_3days.day)
browser.get(new_link)

for i in range(30):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 100000);")

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    T = dt.datetime.strptime(d.find(class_='date').text, '%Y/%m/%d %H:%M')
    
    if T >= pm_3 and T <= pm_5:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
    elif T <= pm_3:
        break

    
```
