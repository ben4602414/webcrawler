### 比較一下範例檔案中的「r.text」與「json.loads(r.text)」讀出來的內容有什麼差異

```python
#r.text
import requests
r = requests.get("http://httpbin.org/get")
print(type(r.text))
r.text
```

```python
#json
import json
print(type(json.loads(r.text)))
json.loads(r.text)
```
