### 尋找一個網站是有作 Header 檢查的（Hint: 有加跟沒有加回傳結果不一樣）

```python
import requests
r = requests.get('https://www.dcard.tw/f/food/p/237446220')
r.text
```

### 用上述網站說明該如何判斷 Header 中必須加上哪些資料才會正確


```python
headers = {'User-Agent': 'Mozilla/5.0'}
r = requests.get('https://www.dcard.tw/f/food/p/237446220', headers = headers)
r.text

```
