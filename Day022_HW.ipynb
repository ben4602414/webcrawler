### 2. 剛剛的做法只能取出第一頁，試著取出所有頁面的資料。

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='/Users/ben/webcrawler/chromedriver')
browser.get("https://hahow.in/courses")
```

```python
# 取得資料（等瀏覽器資料出現後才執行）
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html.parser')
```

```python
# 1. 先取出全部有幾頁
total_pages = 0
for d in soup.find(class_="pagination-container").find_all("li"):
    try:
        x = int(d.find('a').text)
    except:
        continue
        
    if x > total_pages:
        total_pages = x
    else:
        continue
print(total_pages)
```

```python
# 2. 每頁都取出資料放入 courses

courses = []
for i in range(total_pages):
    i += 1
    browser.get("https://hahow.in/courses?page={}".format(i))
    
    for d in soup.find(class_="list-container").find_all('h4'):
        courses.append(d.text)
browser.close()
```

```python
# 3. done!

print(courses)
```
